```python
#Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import os
import folium
import json
```

```python
# This command propts matplotlib visuals to appear in the notebook 
%matplotlib inline
```

```python
path=r'C:\Users\User\Project Covid'
```

```python
#Importing merged data
df_merged_covid = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'covid_merged_datanew1.csv'))
```

```python
country_geo = r'C:\Users\User\Project Covid\02 DATA\Original data\us-states.json'
```

```python
f = open(r'C:\Users\User\Project Covid\02 DATA\Original data\custom.geo.json',errors="ignore")
```

```python
# returns JSON object asa dictionary
data = json.load(f)
```

```python
# Iterating through the json list
for i in data['features']:
     print(i)
```

```python
#Create a list of necessary columns\n",
columns = [ 'state','covid_cases','covid_deaths','Population']
```

```python
#Createa a subset\n",
df_spatial = df_merged_covid[columns]
```

```python
map = folium.Map(location = [40, -95], zoom_start = 4)
folium.Choropleth(
    geo_data = country_geo, 
    data = df_spatial,
    columns = ['state', 'covid_cases'],
    key_on = 'feature.properties.name',
    fill_color = 'YlOrBr', fill_opacity=0.6, line_opacity=0.1,
    legend_name = "cases").add_to(map)
   

folium.LayerControl().add_to(map)


map
```

```python
map.save(os.path.join(path, '04 Analysis', 'Visualizations', 'plot_states_cases.html'))
```
